## **1. Umgang mit fehlenden Werten**

#### Textspalten

```python
# Entscheidung NUll-Werte
# https://www.auto-data.net/de/suzuki-verona-2.5i-156hp-automatic-23901
# Es handelt sich also um einen Benziner

df.loc[df['Engine Fuel Type'].isna(), 'Engine Fuel Type'] = 'regular unleaded'

# Prüfung, ob Übertragung geklappt hat
display(df[df['Engine Fuel Type'].isna()])
```

```python
# NUll-Werte "Market Category"
# In dieser Spalte fehlen zuviele Werte, daher wird für leere Einträge
# "Unknown" eingetragen
df['Market Category'] = df['Market Category'].fillna('Unknown')

# Prüfung
display(df[df['Market Category'].isna()])
```

#### Nummerische Spalten

```python
# NUll-Werte "Engine HP"
display(df[df['Engine HP'].isna()])

# Eigentlich dumme Idee.
#     - ist zwar robust gegenüber Ausreißern nach unten und oben
#     - aber dadurch, dass alle fehlenden den Median bekommen wird die
#         Standardabweichung im Datenset künstlich verringert
#     - unrealistisch Fiat, Nissan und Tesla identische HP?
#     - Korrelationen weniger gut, da Werte Median (Glättung der Korrelation?)
df['Engine HP'] = df['Engine HP'].fillna(df['Engine HP'].median())

# Prüfung - keine Null-Werte mehr vorhanden
display(df[df['Engine HP'].isna()])
```

```python
# NUll-Werte "Engine Cylinders"
display(df[df['Engine Cylinders'].isna()])

# Unique-Werte anzeigen
display(df['Engine Cylinders'].value_counts(dropna=False))

# Elektro auf 0 setzen -> haben keine Zylinder
ev_mask = (df["Engine Fuel Type"] == "electric") & (df["Engine Cylinders"].isna())
df.loc[ev_mask, "Engine Cylinders"] = 0

# Für den Rest der fehlenden Werte nutzen wir den Median
df['Engine Cylinders'] = df['Engine Cylinders'].fillna(df['Engine Cylinders'].median())

# Prüfung
display(df[df['Engine Cylinders'].isna()])
```

```python
# NUll-Werte "Number of Doors"
display(df[df['Number of Doors'].isna()])

# Unique-Werte anzeigen
display(df['Number of Doors'].value_counts(dropna=False))

# Werte recherchiert und eingetragen
# Tesla Model S 2016 -> 4 Türen
df.loc[(df['Make'] == 'Tesla') &
           (df['Model'] == 'Model S') &
           (df['Year'] == 2016) &
           (df['Number of Doors'].isna()), 'Number of Doors'] = 4

# Ferrari FF 2013 -> 3 Türen
df.loc[(df['Make'] == 'Ferrari') &
           (df['Model'] == 'FF') &
           (df['Year'] == 2013) &
           (df['Number of Doors'].isna()), 'Number of Doors'] = 3

# Prüfung
display(df[df['Number of Doors'].isna()])
```

## **2. Datentypen anpassen**


#### Textspalten

```python
# Spalten auswählen
strings_cols = ["Make", "Model", "Engine Fuel Type", "Transmission Type", "Driven_Wheels", "Market Category", "Vehicle Size", "Vehicle Style"]

# Spalten konvertieren
for col in strings_cols:
  df[col] = df[col].astype("string")

# Prüfung
df.info()
```

#### Numerische Spalten

```python
# Konvertierung aller Spalten je nach Daten (INT)
df['Number of Doors'] = df['Number of Doors'].astype('int64')
df['Year'] = df['Year'].astype('int64')

# Konvertierung aller Spalten je nach Daten (FLOAT)
df['Engine Cylinders'] = df['Engine Cylinders'].astype('float64')
df['highway MPG'] = df['highway MPG'].astype('float64')
df['city mpg'] = df['city mpg'].astype('float64')
df['Popularity'] = df['Popularity'].astype('float64')
df['MSRP'] = df['MSRP'].astype('float64')
df['Engine HP'] = df['Engine HP'].astype('float64')

# Prüfung
df.info()
```

## **3. Daten filtern**




```python
# Filtere den Datensatz so, dass nur Autos ab dem Baujahr 1995 enthalten sind.
filtered_df = df[df['Year'] >= 1995]

#Prüfung
display(filtered_df['Year'].unique())
```

## **4. String-Operationen**



```python
# Standardisiere die Einträge in folgenden Spalten, indem du sie in
# Kleinbuchstaben umwandelst:
#     - Vehicle Style
#     - Market Category

# Vorher -> alle eindeutigen Werte -> Market Category
display(filtered_df['Market Category'].unique())

# Konvertierung in Kleinbuchstaben
filtered_df ['Market Category'] = filtered_df['Market Category'].str.lower()

# Prüfung
display(filtered_df['Market Category'].unique())

# -------------------------------------------------------------------

# Vorher -> alle eindeutigen Werte Vehicle Size
display(filtered_df['Vehicle Size'].unique())

# Konvertierung in Kleinbuchstaben
filtered_df ['Vehicle Size'] = filtered_df['Vehicle Size'].str.lower()

# Prüfung
display(filtered_df['Vehicle Size'].unique())
```

# **Feature Engineering**


## **1. Neue Spalten erstellen**

**Total MPG**  ✅
> Erstelle eine neue Spalte als Durchschnitt aus *city mpg* und *highway MPG*.  

**Price per HP**  ✅
> Erstelle eine Spalte berechnet als:  
> *MSRP / Engine HP*

```python
# Benötigte Spalten auswählen
cols = ['highway MPG', 'city mpg']

# Berechnung durchführen und neuer Spalte zuordnen
filtered_df["Total MPG"] = filtered_df[cols].mean(axis = 1)

# Prüfung
display(filtered_df[['highway MPG', 'city mpg', 'Total MPG']])
```

```python
# Benötigte Spalten auswählen
cols = ['MSRP', 'Engine HP']

# Berechnung durchführen und neuer Spalte zuordnen
filtered_df["Price per HP"] = (filtered_df["MSRP"] / filtered_df["Engine HP"]).round(2)

# Prüfung
display(filtered_df[['MSRP', 'Engine HP', 'Price per HP']])
```

## **4. Korrelationsanalyse**  



> Untersuche den Zusammenhang (Korrelation) zwischen folgenden Variablen:  ✅
> - Engine HP  
> - MSRP  
> - Popularity  
> - city mpg  
> - highway MPG  

### **Korrelationsmatrix -> Pearson**

```python
# Berechnet und druckt die Pearson-Korrelationsmatrix,
# die die lineare Beziehung misst.


# Auswahl der benötigten Spalten
cols = ['Engine HP', 'MSRP', 'Popularity', 'city mpg', 'highway MPG']

# Dataframe für die Korrelation
corr_df = filtered_df[cols]

# Korrelationsmatrix
correlation_matrix_pearson = corr_df.corr(method='pearson').round(2)

# Matrix ausgeben
display(correlation_matrix_pearson)

# Plot erstellen
plt.figure(figsize=(8,6))
sns.heatmap(
    correlation_matrix_pearson,
    annot=True,                     # show values
    cmap='coolwarm',                # diverging colormap
    vmin=-1, vmax=1,                # keep scale consistent
    square=True,
    linewidths=0.5
)

plt.title("Correlation Matrix of Key Variables", fontsize=14)
plt.show()
```

### **Korrelationsmatrix -> Spearman**

```python
# Berechnet und druckt die Spearman-Korrelationsmatrix,
# die die monotone Rang-Korrelation misst..


# Auswahl der benötigten Spalten
cols = ['Engine HP', 'MSRP', 'Popularity', 'city mpg', 'highway MPG']

# Dataframe für die Korrelation
corr_df = filtered_df[cols]

# Korrelationsmatrix
correlation_matrix_spearman = corr_df.corr(method='spearman').round(2)

# Matrix ausgeben
display(correlation_matrix_spearman)

# Plot erstellen
plt.figure(figsize=(8,6))
sns.heatmap(
    correlation_matrix_spearman,
    annot=True,                     # show values
    cmap='coolwarm',                # diverging colormap
    vmin=-1, vmax=1,                # keep scale consistent
    square=True,
    linewidths=0.5
)

plt.title("Correlation Matrix of Key Variables", fontsize=14)
plt.show()
```

### **Korrelationsmatrix -> Kendall**

```python
# Berechnet und druckt die Kendall-Korrelationsmatrix,
# die ebenfalls die Rang-Korrelation misst.


# Auswahl der benötigten Spalten
cols = ['Engine HP', 'MSRP', 'Popularity', 'city mpg', 'highway MPG']

# Dataframe für die Korrelation
corr_df = filtered_df[cols]

# Korrelationsmatrix
correlation_matrix_kendall = corr_df.corr(method='kendall').round(2)

# Matrix ausgeben
display(correlation_matrix_kendall)

# Plot erstellen
plt.figure(figsize=(8,6))
sns.heatmap(
    correlation_matrix_kendall,
    annot=True,                     # show values
    cmap='coolwarm',                # diverging colormap
    vmin=-1, vmax=1,                # keep scale consistent
    square=True,
    linewidths=0.5
)

plt.title("Correlation Matrix of Key Variables", fontsize=14)
plt.show()
```
